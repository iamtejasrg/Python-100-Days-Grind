```python
#f(x) is sin(x)/x
#find the maxima of given functon 
#binary search based method for finding x such that slope of f(x) is zero
#the slope changes from positive to negative around the maxima x

def f(x):
    if x!=0:
        return math.sin(x)/x
    else:
        return math.nan;
    
def slopeF(x):
    delta = 0.0001;
    return (f(x+delta)-f(x))/delta

print(slopeF(-0.8))
```

```python
#f(x) is x^2 +sin(x)+ 2x
def f(x):
    if x!=0:
        return (x**2 + math.sin(x) + 2 * x)
    else:
        return math.nan;
    
def slopeF(x):
    delta = 0.001;
    return (f(x+delta)-f(x))/delta

print(slopeF(-2))
```

```python
#importing libraries
import numpy as np
import math
import random

#to find factorial of a given number
def fact(n):
    r=1
    for i in range(1,n+1):
        r *= i
    return r
fact(3)
```

```python
#find a value of pi only using the formula of area and circumperance
#we using montecorlo approach
#area of a square=1
#radius =0.5
#area of the circle=PI*0.5^2 (PI*R^2)

def incircle(x,y):
    if math.sqrt((x-0.5)**2+(y-0.5)**2) <.5:
        return True
    else:
        return False
    
cntIncircle=0
n=100000;
for i in range(n):
    x=random.random()
    y=random.random()
    
    if incircle(x,y):
        cntIncircle+=1
        
print(cntIncircle/n *4)
        
```
